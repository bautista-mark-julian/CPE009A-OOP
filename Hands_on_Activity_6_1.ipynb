```python
import random

ROLES = ["Novice", "Swordsman", "Archer", "Magician"]
MODES = ["Single Player", "Player vs Player"]

class Player:
    def __init__(self, name, role):
        self.name = name
        self.role = role
        self.hp = 100
        self.wins = 0

    def attack(self):
        return random.randint(5, 20)

class Game:
    def __init__(self):
        self.players = []

    def select_mode(self):
        print("Select mode:")
        for i, mode in enumerate(MODES, start=1):
            print(f"{i}. {mode}")
        choice = int(input()) - 1
        if choice == 0:
            self.single_player_mode()
        elif choice == 1:
            self.player_vs_player_mode()

    def single_player_mode(self):
        player_name = input("Enter your name: ")
        player = Player(player_name, "Novice")
        self.players = [player, Player("Monster", "Boss")]
        self.start_match()

    def player_vs_player_mode(self):
        self.players = []
        for i in range(2):
            player_name = input(f"Enter Player {i+1} name: ")
            player = Player(player_name, ROLES[self.select_role("Select role:")])
            self.players.append(player)
        self.start_match()

    def select_role(self, prompt):
        print(prompt)
        for i, role in enumerate(ROLES[1:], start=1):
            print(f"{i}. {role}")
        return int(input()) - 1

    def start_match(self):
        print("Match starts!")
        while all(player.hp > 0 for player in self.players):
            attacker, defender = random.sample(self.players, 2)
            print(f"{attacker.name} ({attacker.role}) attacks {defender.name} ({defender.role})")
            damage = attacker.attack()
            defender.hp -= damage
            print(f"{attacker.name} deals {damage} damage to {defender.name}. {defender.name} has {defender.hp} HP remaining.")
        winner = max(self.players, key=lambda x: x.hp)
        winner.wins += 1
        print(f"{winner.name} wins!")
        self.end_match()

    def end_match(self):
        for player in self.players:
            print(f"{player.name} has {player.wins} wins.")

if __name__ == "__main__":
    game = Game()
    while True:
        game.select_mode()
        play_again = input("Do you want to play again?: ")
        if play_again.lower() != 'y':
            break
```
